```python
#400 elemanlı bir liste için L'i 200 alarak MA hesaplaması yapılacaktır.

import random

N = 400

orgSeq = [0]*N

for i in range(N):
    orgSeq[i] = random.randint(-100,100)

#buraya kadar 200 elemanlı, elemanları -100 ile 100 arasında random değerlenen bir liste üretildi.

deneme=[0]*len(orgSeq)
maList=[0]*len(orgSeq)

t=1

while t<len(orgSeq)+1:
    
    if t<201:

        deneme[t-1]=orgSeq[t-1] #işlem yapmak için, t ilk listenin uzunluğundaki kısmını deneme listesine kopyalıyorum.

        maList[t-1]=sum(deneme)/200
        
        t=t+1
    
    else:
        ma=0
        for i in range(200):
            ma+=orgSeq[t-1-i ]
        maList[t-1]=ma/200
        t=t+1

print("Orijinal Veri")
print(orgSeq)
print("MA'lardan oluşan Liste")
print(maList)
```

```python
import matplotlib.pyplot as plt     
  
plt.figure(figsize=(20, 10))    
  
# Remove the plot frame lines.   
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)       
    
plt.yticks(range(-100, 100, 10), fontsize=14)    
plt.xticks(fontsize=14)    

plt.plot(orgSeq)
plt.plot(maList)
plt.show()
```
